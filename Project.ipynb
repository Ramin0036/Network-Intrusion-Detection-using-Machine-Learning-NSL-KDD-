```python
with open("D:\\ML_PROJECT\\NSL_KDD-master\\Field Names_1.csv") as f:
    columns = [line.strip() for line in f]
test = pd.read_csv("D:\\ML_PROJECT\\NSL_KDD-master\\KDDTest+.csv" , names=columns)
train = pd.read_csv("D:\\ML_PROJECT\\NSL_KDD-master\\KDDTrain+.csv" , names=columns)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
le = LabelEncoder()
scaler = StandardScaler()
```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])
```

```python
counts = data['label'].value_counts()
print(counts)
```

```python
data['label'][data['label'] == 16] = 1
data['label'][data['label'] != 1] = -1
```

```python
data = data.drop(columns='difficulty')
```

```python
data.head()
```

```python
scaler = StandardScaler()
data.iloc[:,:-1] = scaler.fit_transform(data.iloc[:,:-1])
```

```python
data_normal = data[data['label'] == 1]
data_anomaly = data[data['label'] == -1]
```

```python
data_normal.head()
```

```python
data_anomaly.head()
```

```python
data_train = data_normal[:52000]
data_train.shape
```

```python
data_test = np.concat([data_anomaly[:25633] , data_normal[52001 :]])
data_test = pd.DataFrame(data_test , columns=data.columns)
data_test.shape
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
X_train = data_train.drop(columns=['label'])
y_train = data_train['label']
```

```python
X_test = data_test.drop(columns=['label'])
y_test = data_test['label']
```

```python
from sklearn.svm import OneClassSVM
from sklearn.metrics import classification_report
```

```python
model = OneClassSVM(nu= 0.06)
model.fit(X_train)
```

```python
y_pred_train = model.predict(X_train)
```

```python
y_pred_test = model.predict(X_test)
```

```python
print(classification_report(y_pred_train , y_train))
```

```python
print(classification_report(y_pred_test , y_test))
```
